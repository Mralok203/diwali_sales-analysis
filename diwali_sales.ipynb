```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
diwali=pd.read_csv("Diwali Sales Data.csv", encoding= "unicode_escape")
#encoding = "unicode_escape" is use for if open csv file not open andgive some error uncode not defing
diwali.head()
```

```python
diwali.info()
```

```python
diwali.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
diwali
```

```python
diwali.info()
```

```python
diwali.isnull().sum()
```

```python
diwali['Amount'].fillna(diwali['Amount'].mean(),inplace=True)
#diling with missing value replace with mean value in NaN value
```

```python
diwali.isnull().sum()
```

```python
diwali['Amount']=diwali['Amount'].astype('int')
```

```python
diwali.info()
```

```python
diwali.columns
```

```python
diwali.describe()
```

```python
diwali[['Age',"Orders","Amount"]].describe()
```

# eda

```python
gender_ax= sns.countplot(x='Gender', data=diwali)
for bars in gender_ax.containers:
    gender_ax.bar_label(bars)
```

in this graph we know the maximum female buyers are maximum compare to man  

```python
sales_gen= diwali.groupby(['Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x="Gender", y="Amount",data=sales_gen)
```

in this  graph we know maximum number of femal spend maximum money

```python
age_ax= sns.countplot(x='Age Group', data=diwali)
for bars in age_ax.containers:
    age_ax.bar_label(bars)
```

in this graph we know the age between 26-35 are maximum buyers are avilabel

```python
pro_ax= sns.countplot(x='Product_Category', data=diwali,hue="Product_Category")
sns.set(rc={'figure.figsize':(30,30)})
for bars in pro_ax.containers:
    pro_ax.bar_label(bars)
```

in this grap most sales product is clothing and apparelBeauty

```python
diwali.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = diwali.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
